# Let's start with simple plots:

# How would you plot the following data? 
People who signed up for the Python course at ZMT:

    year 1: 10 people
    
    year 2: 24 people
    
    year 3: 18 people
    
    year 4: 120 people

```python
plt.rcParams.update({'font.size': 22}) # Handy line of code to make all the text in your plots as large as you want!


fig, ax = plt.subplots(figsize=(8,8))  # Create a figure.
x = [1, 2, 3, 4] # Create some data.
y = [10, 24, 18, 120]
ax.plot(x,y, label='data', color='tomato', linewidth=8, linestyle='-');  # Plot some data.
ax.set_xlabel('Year')  # Add an x-label to the axes.
ax.set_ylabel('Number of cool people')# Add a y-label to the axes.
ax.set_title("Simple Line Plot")  # Add a title to the axes.
ax.set_xticks([1, 2, 3, 4])
ax.set_xlim(1,4)
plt.show()
```

What if we had to plot a dataset with 200 samples of children's height vs. age?

```python
fig, ax = plt.subplots(figsize=(8,8))  # Create a figure.

x = np.linspace(1,12,200) 
y1 = 0.6+0.05*x+np.random.uniform(0,0.2,200) # Create some data.
y2 = 0.6+0.02*x+np.random.uniform(0,0.2,200)
ax.scatter(x,y1, label='Germany', color='seagreen', s=1e2);  # Plot some data.
ax.scatter(x,y2, label='South Park', color='sandybrown', marker='o', s=1e2);
ax.set_xlabel('Age [yrs]')  # Add an x-label to the axes.
ax.set_ylabel('Height [m]')  # Add a y-label to the axes.
ax.set_title("Simple Scatter Plot")  # Add a title to the axes.
ax.set_ylim(0.8,1.4)
ax.set_xlim(2,12)
ax.legend()  # Add a legend.
plt.show()
```

What if we had to plot a dataset with 10000 samples of height?

```python
fig, ax = plt.subplots(figsize=(8,8))  # Create a figure containing a single axes.
x1 = np.random.normal(1.6, 0.1, 1000) # Create some data
x2 = np.random.normal(1.2, 0.1, 1000) # Create some data
ax.hist(x1, bins=10, range=(1.1,2.1), label='Germany', facecolor='white', edgecolor='seagreen', linewidth=4);  # Plot some data on the axes.
ax.hist(x2, bins=10, range=(0.7,1.7), label='South Park', facecolor='white', edgecolor='sandybrown', linewidth=4, alpha=1);  # Plot some data on the axes.
ax.set_xlabel('Height [m]')  # Add an x-label to the axes.
ax.set_ylabel('Count')  # Add a y-label to the axes.
ax.set_title("Simple Histogram")  # Add a title to the axes.
ax.legend(loc='upper right')  # Add a legend.
ax.set_ylim(0,420)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))  # Create a figure containing a single axes.
data = [x1,x2]
ax.boxplot(data);  # Plot some data on the axes.
ax.set_xlabel('Location')  # Add an x-label to the axes.
ax.set_ylabel('Height [m]')  # Add a y-label to the axes.
ax.set_title("Simple Box Plot")  # Add a title to the axes.
ax.set_xticklabels(['Germany', 'South Park'])
plt.show()
```

# Let's explore the wonderful world of matplolib colors!

```python
fig, ax = plt.subplots(figsize=(8,8)) 
ax.set_title('Blue Circle')
circle1 = plt.Circle((0, 0), 1, color='lightskyblue')
ax.add_patch(circle1)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.show()
```

# And colormaps!

```python
Image(url='https://matplotlib.org/stable/_images/sphx_glr_colormaps_021.png', width=600)
```

```python
fig, ax = plt.subplots(figsize=(8,8)) 

ax.set_title('Funky Circle')
xlist = np.linspace(-1.0, 1.0, 100)
ylist = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
cp = ax.contourf(X, Y, Z, cmap='prism')
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
#fig.colorbar(cp)
plt.show()
# Uncomment the following line to save the figure.
# fig.savefig('Funky_circle.png', transparent=False, dpi=80, bbox_inches="tight")
```
